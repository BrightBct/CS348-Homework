# ขั้นตอนการเตรียมข้อมูล

ทำการ drop แถวที่มีค่า null (บางข้อมูลอาจมีค่าที่เป็น null ซึ่งอาจทำให้ plot กราฟมีปัญหา จึงทำการ drop แถวที่มีค่าเป็น null)

```python
Iris_Flower.dropna(inplace=True)
Red_Wine_Quality.dropna(inplace=True)
Loan_Prediction_train.dropna(inplace=True)
```

## Iris Flower

ทำการ set ขนาดของกราฟเป็น 15, 8

ทำการ set พื้นหลังของกราฟเป็นสีขาว

จากนั้นทำ scatterplot โดยใช้ข้อมูลจาก Iris_Flower โดยแกน x เป็นคอลัมน์ sepal_length และแกน y เป็นคอลัมน์ petal_length และมีการกำหนดสี โดยแบ่งสีตามคอลัมน์ species

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.set_style(style='white')
sns.scatterplot(data = Iris_Flower, x = "sepal_length", y = "petal_length", hue = "species")
```

โดยเลือก 2 คอลัมน์ดังกล่าว เพราะต้องการทราบความแตกต่างระหว่างความยาวของกลีบเลี้ยงและกลีบดอกไม้ จากภาพจะเห็นได้ว่ากลุ่มที่เป็น Iris-setosa จะมีความยาวกลีบเลี้ยงตั้งแต่ 4 จนถึง 6 และมีความยาวกลีบดอกไม้ตั้งแต่ 1 จนถึง 2 และกลุ่มที่เป็น Iris-versicolor จะมีความยาวกลีบเลี้ยงตั้งแต่ 5 จนถึง 7 และมีความยาวกลีบดอกไม้ตั้งแต่ 3 จนถึง 5 และกลุ่มสุดท้ายที่เป็น Iris-virginica จะมีความยาวกลีบเลี้ยงตั้งแต่ 5.5 จนถึง 8 และมีความยาวกลีบดอกไม้ตั้งแต่ 4 จนถึง 7

# Red Wine Quality

ทำการ set ขนาดของกราฟเป็น 15, 8

ทำการ set พื้นหลังของกราฟเป็นสีขาว

จากนั้นทำ scatterplot โดยใช้ข้อมูลจาก Red_Wine_Quality โดยแกน x เป็นคอลัมน์ free sulfur dioxide และแกน y เป็นคอลัมน์ total sulfur dioxide และมีการกำหนดสี โดยแบ่งสีตามคอลัมน์ quality

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.set_style(style='white')
sns.scatterplot(data = Red_Wine_Quality, x = "free sulfur dioxide", y = "total sulfur dioxide", hue = "quality")
```

จะเห็นได้ว่าข้อมูลภายในกลุ่มนั้นมีค่าใกล้เคียงกันมาก ทำให้ไม่เห็นความต่างของข้อมูลทั้ง 2

# Loan Prediction

ทำการ set ขนาดของกราฟเป็น 15, 8

ทำการ set พื้นหลังของกราฟเป็นสีขาว

จากนั้นทำ scatterplot โดยใช้ข้อมูลจาก Loan_Prediction_train โดยแกน x เป็นคอลัมน์ LoanAmount และแกน y เป็นคอลัมน์ Credit_History และมีการกำหนดสี โดยแบ่งสีตามคอลัมน์ Loan_Status

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.set_style(style='white')
sns.scatterplot(data = Loan_Prediction_train, x = "LoanAmount", y = "Credit_History", hue = "Loan_Status")
```

โดยเหตุผลในการเลือก 2 features นี้มาทำ Scatter Plot เพราะอยากทราบว่า feature ไหนที่ส่งผลกระทบต่อการอนุมัติการกู้เงิน ซึ่งคิดว่า Credit_History ส่งผลให้เกิดการอนุมัติการกู้เงิน และเลือก LoanAmount มาเพื่ออยากทราบว่าจำนวนเงินที่จะกู้ส่งผลต่อการพิจารณาการอนุมัติเงินกู้หรือไม่ จากภาพจะเห็นได้ว่ากลุ่มที่กู้ผ่านนั้นต่อให้จำนวนเงินที่จะกู้มากน้อยเพียงใด หากมีเครดิตส่วนใหญ่ก็จะสามารถกู้ผ่านได้ ส่วนกลุ่มที่กู้ไม่ผ่านจะเป็นกลุ่มที่ไม่มีเครดิต
